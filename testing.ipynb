```python
# To obtain skin profile data from discover
# Note: can't seem to obtain all discover entries... why?

discover_single = pd.read_csv('discover_print_to_csv_2.csv')
email_skin_profile =  discover_single[['email', 'request.ageRange', 'request.gender', 'request.sensitivity','request.concerns', 'timestamp']].drop_duplicates(subset=['email'])
email_skin_profile['timestamp'] = pd.to_datetime(email_skin_profile['timestamp'])
email_skin_profile['request.sensitivity'] = email_skin_profile['request.sensitivity'].replace([True, False], [1,0])
email_skin_profile.dropna(inplace = True)

# To convert strings of list into lists
import ast
email_skin_profile['request.concerns'] = email_skin_profile['request.concerns'].apply(ast.literal_eval)


concerns_dummies = pd.get_dummies(email_skin_profile['request.concerns'].apply(pd.Series).stack()).sum(level=0)
email_skin_profile = pd.concat([email_skin_profile, concerns_dummies], axis = 1).drop('request.concerns', axis= 1)
```
